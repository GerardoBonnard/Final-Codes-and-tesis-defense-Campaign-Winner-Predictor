# Estudio Final Campaign Winner Predictor con delimitacion de porcentage de Fraude, comparacion y defensa de tesis 

```python
# a Noboa se le resta el porcentage de fraude y a luisa se le suma su porcentage de fraude
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Candidates
candidates = ['Daniel Noboa']

# Official Results
official_results = [44.17]

# Predicted Results
predicted_results = [54.65]

# Create a bar chart
x = np.arange(len(candidates))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, official_results, width, label='Official Results (Blue)', color='blue')
bars2 = ax.bar(x + width/2, predicted_results, width, label='Predicted Results (Orange)', color='orange')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Percentage (%)')
ax.set_title('Comparison of Daniel Noboa: Official vs Predicted Results')
ax.set_xticks(x)
ax.set_xticklabels(candidates)
ax.legend()

# Add value labels on top of the bars
def add_value_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(round(height, 2)),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

add_value_labels(bars1)
add_value_labels(bars2)

plt.ylim(0, 60)  # Set y-axis limit for better visualization
plt.grid(axis='y')  # Optional: Add grid lines for better readability

# Show the plot
plt.show()
```

# Ahora saquemos el porcentage de aparente  Fraude realizado a daniel Noboa 

```python
# Define the numbers
num1 = 54.65
num2 = 44.17

# Perform the subtraction
result = num1 - num2

# Print the result
print("The result of", num1, "-", num2, "is:", result)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Observed frequencies (official results)
observed = np.array([56.94, 47.15])

# Expected frequencies (predicted results)
expected = np.array([55.62, 44.38])

# Create bar charts for observed and expected frequencies
labels = ['Daniel Noboa', 'Luisa Gonzlez']
x = np.arange(len(labels))  # the label locations

# Set the width of the bars
width = 0.35  

fig, ax = plt.subplots()
bars1 = ax.bar(x + width/2, expected, width, label='Resultados oficiales CNE', color='blue')  # Expected frequencies
bars2 = ax.bar(x - width/2, observed, width, label='Prediccin + expectativa de fraude', color='orange')  # Observed frequencies

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Frequencies')
ax.set_title('Prediccin + expectativa de Fraude vs Resultados Oficiales CNE')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Add value labels on top of the bars
def add_value_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

add_value_labels(bars1)
add_value_labels(bars2)

# Show grid for better readability
ax.yaxis.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Observed frequencies (official results)
observed = np.array([56.94, 47.15])

# Expected frequencies (predicted results)
expected = np.array([55.62, 44.38])

# Calculate correct predictions
correct_predictions = np.sum(np.minimum(observed, expected))

# Calculate total predictions
total_predictions = np.sum(observed)

# Calculate accuracy
accuracy = (correct_predictions / total_predictions) * 100

accuracy
```
